```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.prompts.few_shot import FewShotChatMessagePromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(
    temperature=0.1,
    streaming=True,
    callbacks=[
        StreamingStdOutCallbackHandler(),
    ],
)

examples = [
    {
        "question": "서울",
        "answer": """
            면적 : 605.2 km2\n
            인구수 : 약 940만 명 (Approx. 9.4 Million)\n
            추천 맛집 3곳 :\n
            1. Myeongdong Kyoja (명동교자)\n
            2. Woo Lae Oak (우래옥)\n
            3. Tosokchon Samgyetang (토속촌 삼계탕)\n
            추천 관광지 3곳 :\n
            1. Gyeongbokgung Palace (경복궁)\n
            2. N Seoul Tower (N서울타워)\n
            3. Bukchon Hanok Village (북촌 한옥마을)\n
        """
    }    
]

example_prompt = PromptTemplate.from_template("Human: {question} AI: {answer}")

prompt = FewShotPromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
    suffix="Human: What do you know about {movie}?",
    input_variables=["moview"]
)

chain1 = prompt | chat
chain1.invoke({"movie": "신세계"})


# t = PromptTemplate(
#     template="What is the capital of {country}",
#     input_variables=["country"],
# )

# t.format(country="Korea")

# chat.predict("What do you know about korea?")


# example_template = """
#     Human:{question}
#     AI:{
# """
```
