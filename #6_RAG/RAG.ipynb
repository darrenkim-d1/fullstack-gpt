```python
from langchain.embeddings import OpenAIEmbeddings, CacheBackedEmbeddings
from langchain.document_loaders import UnstructuredFileLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.storage import LocalFileStore

cache_dir = LocalFileStore("../.cache/")

splitter2 = CharacterTextSplitter(
    separator="\n",
    chunk_size=600,
    chunk_overlap=100
)

embedder = OpenAIEmbeddings()

vector = embedder.embed_query("Hi")

len(vector)

loader = UnstructuredFileLoader("../files/chapter_01.pdf")

docs = loader.load_and_split(text_splitter=splitter2)

embeddings = OpenAIEmbeddings()

chaced_embeddings = CacheBackedEmbeddings.from_bytes_store(embeddings, cache_dir)

vectorstore = Chroma.from_documents(docs, chaced_embeddings)


```

```python
result = vectorstore.similarity_search("where dows winston live")

len(result)
result
```
