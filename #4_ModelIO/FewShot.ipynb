```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.prompts.few_shot import FewShotPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(
    temperature=0.1,
    streaming=True,
    callbacks=[
        StreamingStdOutCallbackHandler(),
    ],
)

examples = [
    {
        "question": "What do you know about Dark Knight?",
        "answer": """
            Here is what I know:
            Title: Dark Knight
            Director: Christopher Nolan
            Actors: Christian Bale, Heath Ledger, Aaron Eckhart, Maggie Gyllenhaal
            Budget: $185 million
            Box Office Earnings: $1 billion
            Genre: Action
            Synopsis: Synopsis: The Dark Knight
Batman, Lieutenant Gordon, and District Attorney Harvey Dent form an alliance to dismantle organized crime in Gotham City. (배트맨, 고든 경감, 그리고 검사 하비 덴트는 고담시의 조직범죄를 소탕하기 위해 동맹을 맺습니다.)

However, their efforts are challenged by a mysterious and sadistic criminal mastermind known as the Joker, who seeks to plunge the city into anarchy. (하지만 도시를 무정부 상태로 몰아넣으려는 '조커'라고 알려진 신비롭고 가학적인 범죄의 천재가 나타나 그들의 노력에 도전합니다.)

As the Joker's reign of chaos spreads, Batman must confront a deeply personal conflict that tests everything he stands for. (조커가 일으킨 혼란이 퍼지면서, 배트맨은 자신의 신념을 시험하는 깊은 개인적인 갈등에 직면하게 됩니다.)
        """
    }    
]

example_prompt = PromptTemplate.from_template("Human: {question} AI: {answer}")

prompt = FewShotPromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
    suffix="Human: What do you know about {movie}?",
    input_variables=["moview"]
)

chain1 = prompt | chat
chain1.invoke({"movie": "신세계"})


# t = PromptTemplate(
#     template="What is the capital of {country}",
#     input_variables=["country"],
# )

# t.format(country="Korea")

# chat.predict("What do you know about korea?")


# example_template = """
#     Human:{question}
#     AI:{
# """
```
