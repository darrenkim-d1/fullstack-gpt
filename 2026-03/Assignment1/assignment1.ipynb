```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate, ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(temperature=0.1, streaming=True, callbacks=[StreamingStdOutCallbackHandler()])

template1 = ChatPromptTemplate.from_messages(
    [
        ("system", "너는 시인이야, 전달받은 주제로 시를 지어줘"),
        ("human", "{programLanguage} 언어로 시를 만들어줘")
    ]
)

chain1 = template1 | chat
# chain1.invoke({
#     "programLanguage":"파이썬"
# })

template2 = ChatPromptTemplate.from_messages(
    [
        ("system","너는 시를 설명해주는데 특화되어 있어, 전달된 시를 자세하게 설명해줘, 대답할때에는 기존 내용과 구분될 수 있도록 줄바꿈문자로 시작해줘"),
        ("human", "{param}")
    ]
)
chain2 = template2 | chat

resultChain = {"param": chain1} | chain2
resultChain.invoke({
    "programLanguage": "파이썬"
})
```
