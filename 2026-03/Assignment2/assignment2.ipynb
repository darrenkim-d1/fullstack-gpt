```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts.few_shot import FewShotChatMessagePromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler
from langchain.prompts import ChatMessagePromptTemplate, ChatPromptTemplate


chat = ChatOpenAI(
      temperature=0.1
    , streaming=True
    , callbacks=[
        StreamingStdOutCallbackHandler()
    ]
)

examples = [
    {
        "movie": "인셉션",
        "answer":   """
                감독 : 크리스토퍼 놀란
                주요출연진 : 레오나르도 디카프리오, 와타나베 켄, 조셉 고든 레빗, 마리옹 꼬띠아르
                예산 : 1억 6,000만 달러
                흥행수익 : 8억 3,680만 달러
                영화의장르 : SF, 액션, 스릴러
                시놉시스 : 타인의 꿈속에 침투해 생각을 훔치는 보안 전문가 코브가 반대로 누군가의 머릿속에 정보를 심는 '인셉션' 작전을 의뢰받으며 벌어지는 이야기입니다.
                    """
    },
    {
        "movie": "기생충",
        "answer":   """
                감독 : 봉준호
                주요출연진 : 송강호, 이선균, 조여정, 최우식, 박소담
                예산 : 약 110억 원 (약 850만 달러)
                흥행수익 : 2억 6,300만 달러
                영화의장르 : 드라마, 블랙 코미디, 스릴러
                시놉시스 : 전원 백수인 기택네 장남 기우가 고액 과외 면접을 위해 IT 기업 박 사장네 집에 발을 들이면서 시작된 두 가족의 만남이 걷잡을 수 없는 사건으로 번져가는 과정을 그립니다.
                    """
    },
    {
        "movie": "라라랜드",
        "answer":   """
                감독 : 데미언 셔젤
                주요출연진 : 라이언 고슬링, 엠마 스톤
                예산 : 3,000만 달러
                흥행수익 : 4억 4,890만 달러
                영화의장르 : 뮤지컬, 로맨스, 드라마
                시놉시스 : 꿈을 쫓는 재즈 피아니스트 세바스찬과 배우 지망생 미아가 예술의 도시 LA에서 만나 서로의 열정을 응원하며 사랑에 빠지는 마법 같은 순간을 담았습니다.
                    """
    },
]

example_prompt = ChatPromptTemplate.from_messages(
    [
        ("human", "{movie}에 대한 정보를 찾아줘"),
        ("ai", "{answer}")
    ]
)

example_prompt = FewShotChatMessagePromptTemplate(
    example_prompt=example_prompt,
    examples=examples
)

final_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "너는 영화매니아야, 짧게 대답해줘"),
        example_prompt,
        ("human", "{movie}에 대한 정보를 찾아줘")
    ]
)

chain = final_prompt | chat

chain.invoke({"movie": "아이언맨"})
```
